```r
# filter where the Outcome column equals 1
dplyr::filter(diabete, Outcome == 1)
```

```r
# combine certain rows together
dplyr::distinct(rbind(diabete[5:15,], diabete[105:115,], diabete[205:215,]))
```

```r
# cut certain rows and dispaly them
dplyr::slice(diabete, 30:39)
```

```r
# random select certain rows (by numbers)
dplyr::sample_n(diabete, 9)
```

```r
# random select certain rows (by percentage)
dplyr::sample_frac(diabete, 0.2)
```

```r
# rearrange the database according to a specific column (ascending)
dplyr::arrange(diabete, Insulin)
```

```r
# rearrange the database according to a specific column (descending)
dplyr::arrange(diabete, desc(BloodPressure))
```
